```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt


df= pd.read_csv("phishing_attacks_2.csv")
df['Number of Phishing Attacks'] = df['Number of Phishing Attacks'].str.replace(',', '')

years = np.array(df['Year'])
phishing_attacks = np.array(df['Number of Phishing Attacks'])
years = years.astype(int)
phishing_attacks = phishing_attacks.astype(int)

# Plot the data
#plt.plot(years, phishing_attacks, 'o')
#plt.xlabel('Year')
#plt.ylabel('Number of Phishing Attacks')
#plt.title('Phishing Attacks per Year (2000-2020)')
#plt.show()

# Reshape the data to work with LinearRegression
X = years.reshape(-1, 1)
y = phishing_attacks.reshape(-1, 1)

# Fit the linear regression model
model = LinearRegression().fit(X, y)

# Make predictions for future years
future_years = np.array(range(2021, 2031)).reshape(-1, 1)
predicted_attacks = model.predict(future_years)

# Plot the data and predictions
plt.plot(years, phishing_attacks, label='Actual data')
plt.plot(future_years, predicted_attacks, label='Predicted data')
plt.xlabel('Year')
plt.ylabel('Number of phishing attacks')
plt.title('Phishing attacks over time')
plt.legend()
plt.show()
```
